## Ease of Movement indicators

```python
# EASE OF MOVEMENT INDICATOR for Tesla

import yfinance as yf
import pandas as pd
import numpy as np

Tickers=["tsla"] 

df=yf.download(Tickers, start="2021-05-23",end="2023-05-23")  # time period to be observed

high,low,volume,close=df["High"],df["Low"],df["Volume"],df["Close"]

scale=100000000  #selected according to the volumes

list_emvs=["NaN"]    #list that contains all emvs and it will be calculated from day 2 so 1st day NaN

def dist_moved(high,low,PH,PL):   #PH-prior high, PL-prior low
    dist_moved=((high+low)/2)-((PH+PL)/2)    #function clculating distance moved
    return dist_moved

def box_ratio(high,low,volume):
    box_ratio=(volume/scale)/(high-low)   #function calculating box ratio
    return box_ratio

def one_period_emv(dist_moved,box_ratio):
    one_period_emv=dist_moved/box_ratio    #function calculating one period emv from distance nmoved and box ratio
    return one_period_emv

for i in range(1,len(high)):
    distMoved=dist_moved(high[i],low[i],high[i-1],low[i-1])    #calculation of emvs and storing them in that list created
    boxRatio=box_ratio(high[i],low[i],volume[i])
    oneEmv=one_period_emv(distMoved,boxRatio)
    list_emvs.append(oneEmv)

onePeriod={'one_period':list_emvs} #made this into dict type and then to daataframe as using rolling function
oneperiod=pd.DataFrame(onePeriod)

emv_14=oneperiod.one_period.rolling(14).mean()
EMV=list(emv_14)  #first 13 will be NaN as after 14 day we will get first 14 period EMV

dataframe={'Close':close,'High':high,'Low':low,'One Period EMV':list_emvs,'14 period EMV':EMV}
data=pd.DataFrame(dataframe)
data
```
