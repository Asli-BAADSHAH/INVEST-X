# Momentum Indicators

## Ichimoku Cloud lines

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Tickers=["tsla"]
df=yf.download(Tickers, start="2021-05-23", end="2023-05-23")

def calc_ichimoku(df):
    # Tenkan-sen (Conversion Line)
    df['Conversion Line'] = (df['High'].rolling(window=9).max() +
                        df['Low'].rolling(window=9).min()) / 2

    # Kijun-sen (Base Line)
    df['Base line'] = (df['High'].rolling(window=26).max() +
                       df['Low'].rolling(window=26).min()) / 2

    # Senkou Span A (Leading Span A)
    df['Leading Span A'] = ((df['Conversion Line'] + df['Base line']) / 2).shift(26)

    # Senkou Span B (Leading Span B)
    df['Leading Span B'] = ((df['High'].rolling(window=52).max() +
                            df['Low'].rolling(window=52).min()) / 2).shift(26)

    return df

data=calc_ichimoku(df)
dict={'Conversion Line':data['Conversion Line'],'Base line':data['Base line'],'Leading Span A':data['Leading Span A'],'Leading Span B':data['Leading Span B']}
ichimoku=pd.DataFrame(dict)
ichimoku
```
