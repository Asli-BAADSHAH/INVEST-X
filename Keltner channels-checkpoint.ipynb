# Volatility Indicator

## Keltner channel

```python
# calculating middle line, lower and upper bands for tesla.inc

import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st

Tickers=["tsla"]

df_for_first_ema=yf.download(Tickers,start="2021-03-28",end="2021-04-27")  #20 prior day data 

df=yf.download(Tickers, start="2021-05-23",end="2023-05-23")   # we'll calculate things for this time frame

adj_close,high,low=df['Adj Close'],df['High'],df['Low']

EMA=[st.mean(df_for_first_ema['Adj Close'])]      #list that will contain EMA's for whole time period
# since we don't have any prior ema so simple moving average of prev adj close will be the first ema

def true_range(high,low,close_y):  #function calculating true range 
    tr=max(abs(high-low),abs(high-close_y),abs(low-close_y))
    return tr

def atr_calc(tr,prev_atr):         #function calculating average true range
    atr=(prev_atr*(20-1)+tr)/20    # 20 taken as calculating 20 period atr
    return atr

def ema_calc(close,close_y):   #function to calculate ema's
    multiplier=2/21
    ema=close*multiplier+close_y*(1-multiplier)
    return ema

tr_for_27=[]  #list that will store prior 20 true ranges before our time period

for i in range(1,len(df_for_first_ema['Adj Close'])):  
    tr_for_27.append(true_range(df_for_first_ema['High'][i],df_for_first_ema['Low'][i],df_for_first_ema['Adj Close'][i-1]))

avg_true_range=[st.mean(tr_for_27)]  #1st 20 period average true range will be mean of prior 20 true ranges
    
for i in range (1,len(adj_close)):     #ema for all days in the time period will be calculated 
    EMA.append(ema_calc(adj_close[i],adj_close[i-1]))

for i in range (1,len(adj_close)):   #20 period atr's will be calculated and stored in list
    avg_true_range.append(atr_calc(true_range(high[i],low[i],adj_close[i-1]),avg_true_range[i-1]))
    
ema=np.array(EMA)
atr=np.array(avg_true_range)
upper_band=list(ema+2*atr)
lower_band=list(ema-2*atr)

k_c={'Adj Close':adj_close,'Lower Band':lower_band,'Middle Line':EMA,'Upper Band':upper_band}
keltner_channel=pd.DataFrame(k_c)  #presenting it in form of a dataframe
keltner_channel
```

```python
#showing keltner channels in a plot
plt.figure(figsize=(12,6))
plt.plot(adj_close, linewidth = 2, label = 'Adjusted close')
plt.plot(keltner_channel['Upper Band'], linewidth = 2, color = 'orange', linestyle = '--', label = 'KC UPPER 20')
plt.plot(keltner_channel['Middle Line'], linewidth = 1.5, color = 'grey', label = 'KC MIDDLE 20')
plt.plot(keltner_channel['Lower Band'], linewidth = 2, color = 'red', linestyle = '--', label = 'KC LOWER 20')
plt.legend(loc = 'lower right')
plt.title('INTC KELTNER CHANNEL 20 TRADING SIGNALS')
```
